```python
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
```

```python
#data trasformation
```

```python
df["Tax"] = df["Salary"] * 0.05
```

```python
df
```

```python
df["City"] = df["City"].str.lower()
```

```python
df
```

```python
df["Experience"] = [2, 5, 4, 3, 1, 5]
```

```python
def salary_category(salary):
    if salary < 50000:
        return "Low"
    elif salary <= 60000:
        return "Medium"
    else:
        return "High"

df["SalaryCategory"] = df["Salary"].apply(salary_category)
```

```python
df["Name"] = df["Name"].str.upper()
```
