```python
import pandas as pd

df = pd.DataFrame({
    'Department':['IT','IT','HR','HR','Finance','Finance'],
    'Employee':['A','B','C','D','E','F'],
    'Salary':[60000,70000,50000,55000,80000,90000],
    'Experience':[2,4,3,5,6,8]
})

print(df)
```

```python
avg_salary = df.groupby('Department')['Salary'].mean()
print(avg_salary)
```

```python
max_experience = df.groupby('Department')['Experience'].max()
print(max_experience)
```

```python
total_salary = df.groupby('Department')['Salary'].sum()
print(total_salary)
```

```python
employee_count = df.groupby('Department')['Employee'].count()
print(employee_count)
```

```python
avg_salary_exp = df.groupby('Department')[['Salary', 'Experience']].mean()
print(avg_salary_exp)
```

```python
summary = df.groupby('Department').agg({
    'Salary': ['mean', 'sum'],
    'Experience': ['mean', 'max'],
    'Employee': 'count'
})

print(summary)
```

```python
employees = pd.DataFrame({
    'Emp_ID':[1,2,3,4],
    'Name':['A','B','C','D']
})

departments = pd.DataFrame({
    'Emp_ID':[1,2,4,5],
    'Department':['IT','HR','Finance','Sales']
})
```

```python
inner_merge = pd.merge(employees, departments, on='Emp_ID', how='inner')
print(inner_merge)
```

```python
left_merge = pd.merge(employees, departments, on='Emp_ID', how='left')
print(left_merge)
```

```python
right_merge = pd.merge(employees, departments, on='Emp_ID', how='right')
print(right_merge)
```

```python
outer_merge = pd.merge(employees, departments, on='Emp_ID', how='outer')
print(outer_merge)
```

```python
result = pd.merge(employees, departments, on='Emp_ID', how='left')

employees_without_dept = result[result['Department'].isna()]
print(employees_without_dept)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'Name':['A','B'],
    'Salary':[50000,60000]
})

df2 = pd.DataFrame({
    'Name':['C','D'],
    'Salary':[70000,80000]
})
```

```python
row_concat = pd.concat([df1, df2])
print(row_concat)
```

```python
col_concat = pd.concat([df1, df2], axis=1)
print(col_concat)
```

```python
row_concat_reset = pd.concat([df1, df2], ignore_index=True)
print(row_concat_reset)
```

```python
concat_keys = pd.concat([df1, df2], keys=['DF1', 'DF2'])
print(concat_keys)
```

```python
import pandas as pd

sales = pd.DataFrame({
    'Region':['East','East','West','West','North'],
    'Product':['A','B','A','B','A'],
    'Sales':[100,150,200,250,300]
})

print(sales)
```

```python
total_sales_region = pd.pivot_table(
    sales,
    values='Sales',
    index='Region',
    aggfunc='sum'
)

print(total_sales_region)
```

```python
sales_region_product = pd.pivot_table(
    sales,
    values='Sales',
    index='Region',
    columns='Product',
    aggfunc='sum'
)

print(sales_region_product)
```

```python
dates = pd.date_range(
    start='2025-01-01',
    periods=10,
    freq='D'
)

ts = pd.DataFrame({
    'Date':dates,
    'Sales':[100,120,130,110,150,160,170,180,175,190]
})
```

```python
ts['Date'] = pd.to_datetime(ts['Date'])

print(ts.dtypes)
```

```python
ts['Year'] = ts.index.year
ts['Month'] = ts.index.month
ts['Day'] = ts.index.day

print(ts.head())
```
