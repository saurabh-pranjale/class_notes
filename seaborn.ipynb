```python
#Identify outliers from the plots

Q1 = df['Attendance_Percentage'].quantile(0.25)
Q3 = df['Attendance_Percentage'].quantile(0.75)
IQR = Q3 - Q1

outliers_attendance = df[
    (df['Attendance_Percentage'] < Q1 - 1.5*IQR) |
    (df['Attendance_Percentage'] > Q3 + 1.5*IQR)
]

print(outliers_attendance)
```
