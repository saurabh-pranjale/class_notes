```python
import matplotlib.pyplot as plt
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 150, 180, 170, 210, 250]
```

```python
plt.plot(months, sales)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
plt.plot(months, sales, color='red', marker='o')
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
products = ["Laptop", "Phone", "Tablet", "Watch"]
units_sold = [45, 80, 30, 25]
```

```python
plt.bar(products, units_sold)
plt.title("Units Sold")
plt.xlabel("Products")
plt.ylabel("Units")
plt.show()
```

```python
bars = plt.bar(products, units_sold, color='orange')

for bar in bars:
    plt.text(bar.get_x(), bar.get_height(), bar.get_height())

plt.xticks(rotation=45)

plt.title("Units Sold")
plt.xlabel("Products")
plt.ylabel("Units")
plt.show()
```

```python
plt.pie(units_sold, labels=products)
plt.show()
```

```python
plt.pie(
    units_sold,
    labels=products,
    autopct='%1.1f%%',
    explode=[0.1, 0, 0, 0],  # Explodes first slice
    shadow=True
)

plt.show()
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [45, 50, 55, 60, 68, 75, 85, 92]
```

```python
plt.scatter(study_hours, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

```python
plt.scatter(study_hours, marks, color='red', s=100)

plt.plot(study_hours, marks, color='blue')  # simple trend line

plt.title("Study Hours vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

```python

# ---------------- Sample Data ----------------
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
attendance = [85, 90, 88, 92, 95, 93]

subjects = ["Math", "Science", "English", "History"]
marks = [78, 85, 88, 74]

study_hours = [1, 2, 3, 4, 5, 6]
marks_scatter = [50, 55, 65, 70, 85, 90]

activities = ["Study", "Play", "Sleep", "Social Media"]
activity_time = [6, 2, 8, 4]

# ---------------- Dashboard ----------------
plt.figure(figsize=(10, 8))

# Chart 1: Line Chart - Monthly Attendance
plt.subplot(2, 2, 1)
plt.plot(months, attendance, marker='o')
plt.title("Monthly Attendance")

# Chart 2: Bar Chart - Subject Marks
plt.subplot(2, 2, 2)
plt.bar(subjects, marks, color='orange')
plt.title("Subject Marks")

# Chart 3: Scatter Plot - Study Hours vs Marks
plt.subplot(2, 2, 3)
plt.scatter(study_hours, marks_scatter, color='red')
plt.title("Study Hours vs Marks")

# Chart 4: Pie Chart - Time Spent on Activities
plt.subplot(2, 2, 4)
plt.pie(activity_time, labels=activities, autopct='%1.1f%%')
plt.title("Time Spent on Activities")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# ---------------- Sample Data ----------------
movies = ["Pathaan", "Jawan", "Dangal", "PK", "3 Idiots"]
box_office = [1050, 1100, 2000, 800, 700]  # in crores

ratings = [7.0, 7.5, 8.4, 8.1, 8.4]

years = [2010, 2014, 2016, 2019, 2023]
budget = [120, 150, 90, 200, 250]

genres = ["Action", "Action", "Sports", "Comedy", "Drama"]
genre_count = [2, 2, 1, 1, 1]

# ---------------- Dashboard ----------------
plt.figure(figsize=(10, 8))

# Chart 1: Line Chart - Box Office Trend
plt.subplot(2, 2, 1)
plt.plot(movies, box_office, marker='o', color='blue')
plt.title("Box Office Collection")

# Chart 2: Bar Chart - Movie Ratings
plt.subplot(2, 2, 2)
plt.bar(movies, ratings, color='orange')
plt.title("IMDb Ratings")

# Chart 3: Scatter Plot - Budget vs Box Office
plt.subplot(2, 2, 3)
plt.scatter(budget, box_office, color='red')
plt.title("Budget vs Box Office")

# Chart 4: Pie Chart - Genre Distribution
plt.subplot(2, 2, 4)
plt.pie(genre_count, labels=genres, autopct='%1.1f%%')
plt.title("Genre Distribution")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV file
df = pd.read_csv("restaurant_sales_data.csv.csv")

# ---------------- Dashboard ----------------
plt.figure(figsize=(12, 8))

# 1️⃣ Orders per Item
plt.subplot(2, 2, 1)
plt.bar(df["Item"], df["Orders"], color="orange")
plt.title("Orders per Item")
plt.xticks(rotation=45)

# 2️⃣ Revenue per Item
plt.subplot(2, 2, 2)
plt.bar(df["Item"], df["Revenue"], color="green")
plt.title("Revenue per Item")
plt.xticks(rotation=45)

# 3️⃣ Orders vs Revenue (Scatter Plot)
plt.subplot(2, 2, 3)
plt.scatter(df["Orders"], df["Revenue"], color="red")
plt.title("Orders vs Revenue")
plt.xlabel("Orders")
plt.ylabel("Revenue")

# 4️⃣ Top 5 Revenue Items
top5 = df.sort_values("Revenue", ascending=False).head(5)

plt.subplot(2, 2, 4)
plt.bar(top5["Item"], top5["Revenue"], color="blue")
plt.title("Top 5 Revenue Items")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
